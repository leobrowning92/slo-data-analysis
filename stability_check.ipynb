```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
y=(np.random.rand(1,100))[0]
x=np.arange(0,10,0.1)
```

```python
fig=plt.figure(figsize=(5,5))
ax = plt.subplot(1,1, 1)


def plot_stability_variations(ax,x,y):
    mean,max,min,std=np.mean(y),np.max(y),np.min(y),np.std(y)
    ax.plot(x,y)
    ax.axhline(mean,color="g")
    ax.axhline(mean+std,color="r")
    ax.axhline(mean-std,color="r")
    return mean,std,max,min

plot_stability_variations(ax,x,y)
```
